# Comparing True Shooting Percentages in Very High and Medium Leverage Situations Between Jayson Tatum and Other Star Players

## The Data Used

All of the data used is from the PBP Stats website - https://www.pbpstats.com/ 

For this project, I used scoring data for several teams during the 2023 - 2024 regular season. Data was obtained for both medium and very high leverage situations. 

the term "leverage" is a term used on the PBP Stats website for describing how much a poessession impacts win probability. These high leverage poessessions are ones where there is the greatest amount of pressure on the players, since they are so important to winning the game. PBP Stats describes some game conditions and their respective leverage states:
- Up 8 with 1:20 left in the 4th: Low
- Up 13 with 3:15 left in the 4th: Low
- Down 24 with 2:18 left in the 3rd: Low
- Up 7 with 2:26 left in the 4th: Medium
- Up 1 with 10:18 left in the 4th: Medium
- Down 4 with 1:29 left in the 4th: High
- Tie game with 6:38 left in the 4th: High
- Down 5 with 7:06 left in the 4th: High
- Down 2 with 0:41 left in the 4th: Very High
- Down 1 with 1:02 left in the 4th: Very High
- Down 2 with 1:36 left in the 4th: Very High

Another term often used throughout this project will be "true shooting". This term represents a statistic which is used to measure how efficient a player is at shooting the basketball. Field goal percent only measures how many times you were able to score the basketball against how many shots you took. However, true shooting also includes any potential free throws you may have made.

## The Libraries Used

For this project, I used Pandas, NumPy, Seaborn and Matplotlib. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
veryhigh_leverage = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/VeryHigh_Leverage_Stats.csv', index_col = 'Name')
# Importing the data for the Boston Celtics in very high leverage situations. 
```

```python
sorted_veryhigh_leverage = veryhigh_leverage.sort_values('Minutes', axis = 0, ascending = False).head(6)
# I only care about scoring data for the top 6 members. In order to get this, I sorted the data by minutes played, and took the top 6 results. 
```

```python

sorted_veryhigh_leverage['FGA'] = sorted_veryhigh_leverage['FG2A'] + sorted_veryhigh_leverage['FG3A']
# The data from PBP Stats does not seem to include total field goal attempts, so I had to make my own by adding 3 point and 2 point attempts. 

sorted_veryhigh_leverage
```

```python
labels = sorted_veryhigh_leverage.index # labels to use for the pie chart. 
plt.pie(data = sorted_veryhigh_leverage, x = 'FGA', colors = sns.color_palette('bright'), autopct = '%0.1f%%', labels = labels)
plt.title('Percent of team shots in high leverage situations (BOS)')
```

As we can see from this pie chart, Jayson Tatum is taking the most shots in high leverage situations on the Boston Celtics. 

```python
sns.barplot(x = sorted_veryhigh_leverage.index, y = sorted_veryhigh_leverage['FGA']) # Barplot to see exactly how many shots are being taken. 
plt.xticks(rotation = 30)
plt.title('Number of shots per player in very high leverage situations (BOS)')
```

```python
sns.barplot(data = sorted_veryhigh_leverage, x = 'Name', y = 'TsPct') # Barplot to examine the true shooting per player. 
plt.xticks(rotation = 30)
plt.title('True shooting percent per player in very high leverage situations (BOS)')
```

## How Does Jayson Tatum Compare to Stephen Curry?

After similar analysis for the Golden State Warriors 2023 - 2024 season, we can see that Stephen Curry takes an even larger percent of his teams shots during very high leverage situations. He has also taken over 10 more shots than Jayson Tatum, and has a significantly higher true shooting percentage. 

Based on this analysis, I think it's fair to say Stephen Curry is more clutch than Jayson Tatum. However, there was never really any doubt in my mind that this would be the case. Just because Jayson Tatum isn't the clutch player of the year doesn't mean that he isn't a clutch player. 

To further investigate this, I obtained the shooting data for 8 more players who are perceived as being clutch. 

```python
Chi_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/CHI_VeryHigh_Leverage.csv', index_col = 'Name') # I need to obtain the data of the teams for each of the players I am looking to compare. 
Okc_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/OKC_VeryHigh_Leverage.csv', index_col = 'Name')
Den_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/DEN_VeryHigh_Leverage.csv', index_col = 'Name')
Nyk_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/NYK_VeryHigh_Leverage.csv', index_col = 'Name')
Dal_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/DAL_VeryHigh_Leverage.csv', index_col = 'Name')
Min_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/MIN_VeryHigh_Leverage.csv', index_col = 'Name')
Lal_stats = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/LAL_VeryHigh_Leverage.csv', index_col = 'Name')
```

```python
Chi_stats['FGA'] = Chi_stats['FG2A'] + Chi_stats['FG3A']
Okc_stats['FGA'] = Okc_stats['FG2A'] + Okc_stats['FG3A']
Den_stats['FGA'] = Den_stats['FG2A'] + Den_stats['FG3A'] # Again, creating a FGA column. 
Nyk_stats['FGA'] = Nyk_stats['FG2A'] + Nyk_stats['FG3A']
Dal_stats['FGA'] = Dal_stats['FG2A'] + Dal_stats['FG3A']
Min_stats['FGA'] = Min_stats['FG2A'] + Min_stats['FG3A']# I am not interested in other players stats relative to their own team for this part, so I am not going to filter out the top 6 players.
Lal_stats['FGA'] = Lal_stats['FG2A'] + Lal_stats['FG3A']
merged_totals = pd.concat([Chi_stats, Okc_stats, Den_stats, Nyk_stats, Dal_stats, Min_stats, Lal_stats, sorted_veryhigh_leverage])  
```

```python
TopTen_Clutch = Warriors_VeryHigh_Leverage.loc[['Stephen Curry']] # Creating a new dataframe to store each players stats for very high leverage situations. 
player_names = ['DeMar DeRozan', 'Shai Gilgeous-Alexander', 'Nikola Jokic', 'Jalen Brunson', 'Luka Doncic', 'Kyrie Irving', 'Anthony Edwards', 'Jayson Tatum', 'LeBron James'] # The players I will be compaing. 

for name in player_names: # For loop to extract each players stats and store them in the dataframe. 
    if name in merged_totals.index:
        temp = merged_totals.loc[[name]]
        TopTen_Clutch = pd.concat([TopTen_Clutch, temp])

TopTen_Clutch
```

```python
sns.barplot(data = TopTen_Clutch, x = 'Name', y = 'TsPct') # Barplot to examine the true shooting for the 10 players. 
plt.xticks(rotation = 75)
plt.title('True shooting percent per player in very high leverage situations')
```

```python
plot = sns.scatterplot(data = TopTen_Clutch, x = 'FGA', y = 'TsPct', hue = 'Name', legend = False) # Scatterplot to examine FGA true shooting across the 10 players. 
plt.title('True shooting percentage with associated shot attempts per player')

for point in range(0, TopTen_Clutch.shape[0]): # For loop to label each point with the player name, rather than using a legend. 
    plot.text(TopTen_Clutch.FGA[point] + 0.5, TopTen_Clutch.TsPct[point] + 0.005, TopTen_Clutch.index[point], horizontalalignment = 'center', size = 'x-small', color = 'black', weight = 'light', rotation = -10)
```

As we can see from the scatterplot, Jayson Tatum has the lowest true shooting percentage of the 10 players sampled. In the official NBA Clutch Player of the Year rankings, he was listed at #9, above LeBron James. Because of this, I would expect everyone else on this list to have better true shooting percentages than Jayson Tatum. However, Jalen Brunson is pretty close to Jayson Tatum on this plot while also being listed at #5 for the Clutch Player of the Year award. To investigate this further, scoring data from medium leverage situations will be compared to the very high leverage sitations.  

```python
TsPct_Diff = pd.DataFrame({'TsPct_VeryHigh' : TopTen_Clutch['TsPct']}) # Creating a new dataframe to hold compare medium and very high leverage situation data. 
TsPct_Diff
```

```python
Chi_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/CHI_Medium_Leverage.csv', index_col = 'Name')
Okc_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/OKC_Medium_Leverage.csv', index_col = 'Name')
Den_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/DEN_Medium_Leverage.csv', index_col = 'Name')
Nyk_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/NYK_Medium_Leverage.csv', index_col = 'Name')
Dal_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/DAL_Medium_Leverage.csv', index_col = 'Name') # Importing the medium leverage data for each team of the players I am examining. 
Min_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/MIN_Medium_Leverage.csv', index_col = 'Name')
Lal_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/LAL_Medium_Leverage.csv', index_col = 'Name')
Bos_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/BOS_Medium_Leverage.csv', index_col = 'Name')
Gsw_stats_medium = pd.read_csv('~/Documents/Jayson_Tatum_Clutch/GSW_Medium_Leverage.csv', index_col = 'Name')
```

```python
df_names = [Chi_stats_medium, Okc_stats_medium, Den_stats_medium, Nyk_stats_medium, Dal_stats_medium, Min_stats_medium, Lal_stats_medium, Bos_stats_medium, Gsw_stats_medium] # List of the medium leverage dataframes
player_names = ['DeMar DeRozan', 'Shai Gilgeous-Alexander', 'Nikola Jokic', 'Jalen Brunson', 'Luka Doncic', 'Kyrie Irving', 'Anthony Edwards', 'Jayson Tatum', 'LeBron James']
merged_medium = pd.concat(df_names) # Merging all medium leverage dataframes into one for easier handling. 
TopTen_Medium = Gsw_stats_medium.loc[['Stephen Curry']] # Creating a new dataframe to hold information for the 10 players I am interested in. 
for name in player_names:
    if name in merged_medium.index: # Filling the newly created dataframe. 
        temp = merged_medium.loc[[name]]
        TopTen_Medium = pd.concat([TopTen_Medium, temp])
        
TopTen_Medium
```

```python
TsPct_Diff['TsPCT_Medium'] = TopTen_Medium['TsPct'] # Adding in a new column for the true shooting in medium leverage situations. 
TsPct_Diff['Diff_Medium_VeryHigh'] = TsPct_Diff['TsPct_VeryHigh'] - TsPct_Diff['TsPCT_Medium'] # New column to show the difference between very high and medium leverage situations true shooting. 
TsPct_Diff['Percent_Change'] = TsPct_Diff['Diff_Medium_VeryHigh'] / TsPct_Diff['TsPCT_Medium'] * 100 # New column to show the percent change from medium leverage to very high leverage situations. 
TsPct_Diff
```

```python
barplt = sns.barplot(x = TsPct_Diff.index, y = TsPct_Diff['Percent_Change']) # Barplot to visualize the percent change. 
plt.xticks(rotation = 90)
plt.ylabel('Percent change')
plt.title('Percent change in true shooting from medium to high leverage situations')
barplt.bar_label(barplt.containers[0]) # Show the percent values on the bars. 
```

## Conclusions 

As shown in the above barplot, Jayson Tatum have the worst drop off in true shooting percentage from medium to very high leverage situations. It seems reasonable to expect a worse shooting perentage in situations with a lot of pressure, but a 45% drop off seems very extreme. Compared to a player like Luka Doncic, who has been known to perform in the clutch, Jayson Tatum does not seem to be clutch. In the case of both Luka Doncic and Kyrie Irving, there is a small sample size, so their true shooting numbers would most likely lower with an increased sample size. However, I would not expect them to lower so much that they would be comparable to Jayson Tatum. 

In conclusion, I think it is fair to say Jayson Tatum is not a particularly clutch player. From a fans perspective, watching him in clutch situations often makes me more than nervous than watching other players. Statistically speaking, his true shooting numbers in very high leverage situations are the lowest of the 10 players examined. Additionally, he has an ~45% drop off in true shooting from medium to very high leverage situations. In an ideal world, the sample sizes between medium leverage and very high leverage situations would be the same in order to give a better comparison, but very high leverage situations are by definition more rare than medium leverages ones.
